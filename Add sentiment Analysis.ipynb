```python
import pandas as pd
df= pd.read_csv("./data/FinalDataSet.csv")
```

```python
df[["Contest","Round", "Semi Final Position",
       "Semi Final Points", "Final Position", "Final Points"]]
```

```python
#Copy the DNQ to the Round
df.loc[(df['Final Position']=='DNQ'),'Round']='DNQ'

df[["Round","Contest","Semi Final Position",
       "Semi Final Points", "Final Position", "Final Points"]].loc[df['Round']=='DNQ']
```

```python
#Remove the non numbers from the positions and points
df["Semi Final Position"] = df["Semi Final Position"].str.replace('[^\d]', '')
df["Semi Final Points"] = df["Semi Final Points"].str.replace('[^\d]', '')
df["Final Position"] = df["Final Position"].str.replace('[^\d]', '')
df["Final Points"] = df["Final Points"].str.replace('[^\d]', '')

# Populate the missing translation lyrics with the local lyrics where they are missing
df['Translation Multi Line'].fillna(df['Local Lyrics Lines'],inplace=True)
# Populate the missing translation lyrics with the local lyrics where they are missing
df['Translation'].fillna(df['Local Lyrics Single Line'],inplace=True)


df["Translation Length"] = df["Translation Multi Line"].str.len()
```

```python
from textblob import TextBlob
df['song_polarity'] = df['Translation'].apply(lambda r: TextBlob(r).sentiment.polarity)
df['song_subjectivity'] = df['Translation'].apply(lambda r: TextBlob(r).sentiment.subjectivity)




```

```python
#Write the processed results out to a new file
df.to_csv('./data/Processed1DataSet.csv')
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
import nltk

#Clear any Sentiment cols
cols = [c for c in df.columns if c.lower()[:4] != 'Sentiment_']
df=df[cols]

foo=""
def f1(foo=foo): return iter(foo.splitlines())

#Loop through the dataframe and add sentimate per line to additional columns in the dataframe
for index, row in df.iterrows():
    #print("======="+row["Song Title"])
    #print(row["Translation Multi Line"])
    
    #corpusReader = nltk.corpus.PlaintextCorpusReader(row["Translation Multi Line"])
    
    
    lines = f1(row["Translation Multi Line"])
    for index2, line in enumerate(lines):
        #tokens = tokenizer.tokenize(line)
        #wordcount =  tokens.count
        #print(f"--{line}{wordcount}")
        new_col_pol = "Sentiment_pol_"+str(index2)
        new_col_suj = "Sentiment_suj_"+str(index2)
        sentpol = TextBlob(line).sentiment.polarity
        sentsuj = TextBlob(line).sentiment.subjectivity
        #print(row[new_col])
        df.set_value(index,new_col_pol,sentpol)
        df.set_value(index,new_col_suj,sentsuj)
        
#Show the shape of the data frame
df.shape    
```

```python
#Write the processed results out to a new file
df.to_csv('./data/Processed1DataSet.csv')
```
